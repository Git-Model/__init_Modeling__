### 2.1 çº¿æ€§æ–¹ç¨‹ç»„ä¸å‘é‡
<div class="alert alert-info" role="alert">
â“ GitModel å…¬å¸çš„å®ä¹ ç”ŸGingerè‡ªå·±åœ¨å®¶ä¸­åé™¢ç»„å»ºäº†ä¸€ä¸ªæ™ºæ…§å†œåœºï¼Œå…»äº†å‡ åªå…¬é¸¡å’Œå…”å­ï¼Œç”±äºåˆšæ¥å®ä¹ ï¼Œè¿˜ä¸ä¼šå¾ˆæ·±çš„æ¨¡å‹ï¼Œé‡‡ç”¨çš„æ‘„åƒå¤´å¯¹åŠ¨ç‰©è®¡æ•°æ—¶å€™ï¼Œåªèƒ½ç»Ÿè®¡æœ‰å¤šå°‘ä¸ªå¤´ï¼Œæœ‰å¤šå°‘åªè„š.ç°åœ¨æ‘„åƒå¤´é‡‡é›†åˆ°çš„æ•°æ®æ˜¯ä¸€å…±æœ‰10ä¸ªå¤´ï¼Œ28åªè„šï¼Œè¯·é—®é¸¡å…”å„æœ‰å‡ åªï¼Ÿ

æˆ‘ä»¬å¯ä»¥åˆ—å‡ºä¸€ä¸ªäºŒå…ƒä¸€æ¬¡æ–¹ç¨‹ï¼Œå‡è®¾é¸¡æœ‰$x$åªï¼Œå…”å­æœ‰$y$åªé‚£ä¹ˆå¯ä»¥å¾—åˆ°

#### 2.1.1 çº¿æ€§æ–¹ç¨‹ç»„
$(\ref{eq1})$åœ¨ä¸­å­¦é˜¶æ®µå«åšäºŒå…ƒä¸€æ¬¡æ–¹ç¨‹ç»„ï¼Œå½“ç„¶è§£è¿™ä¸ªæ–¹ç¨‹ç»„å¾ˆç®€å•ï¼Œç­”æ¡ˆæ˜¯ï¼š$\left\{\begin{array}{l}x=6 \\ y=4\end{array}\right.$.
ç°åœ¨æˆ‘ä»¬å¿˜è®°ä¸­å­¦é˜¶æ®µè§£äºŒå…ƒä¸€æ¬¡æ–¹ç¨‹ç»„çš„é«˜æ–¯æ¶ˆå…ƒæ³•ï¼ŒæŠŠäºŒå…ƒä¸€æ¬¡æ–¹ç¨‹ç»„çš„æœ¬è´¨æŠ½è±¡å‡ºæ¥ï¼Œæ¢è®¨ä¸€ç§è§£å†³äºŒå…ƒä¸€æ¬¡æ–¹ç¨‹ç»„çš„æ–¹æ³•ã€‚é€šè¿‡è§‚å¯Ÿï¼Œæˆ‘ä»¬å‘ç°ï¼šäºŒå…ƒä¸€æ¬¡æ–¹ç¨‹ç»„çš„æœªçŸ¥æ•°çš„é˜¶æ•°éƒ½æ˜¯ä¸€æ¬¡ï¼Œå¼å­ä¸¤è¾¹éƒ½æ˜¯ç”¨ç­‰å¼ç›¸è¿ï¼Œå› æ­¤æˆ‘ä»¬ç»™è¿™ä¸ªæ–¹ç¨‹ç»„èµ·å¦ä¸€ä¸ªåå­—ï¼š**çº¿æ€§æ–¹ç¨‹ç»„**ã€‚åœ¨å‰é¢çš„çº¿æ€§æ–¹ç¨‹ç»„ä¸­ï¼Œäº‹å®ä¸Šå¯ä»¥çœ‹æˆæ˜¯ä¸€å¼ è¡¨æ ¼ï¼Œ

| $x$çš„ç³»æ•° | $y$çš„ç³»æ•° | ç»“æœ |
|:---------:|:---------:|:----:|
|     1     |     1     |  10  |
|     2     |     4     |  28  |

æˆ‘ä»¬æŠŠåŒä¸€ä¸ªæœªçŸ¥æ•°çš„ç³»æ•°æ”¾åœ¨ä¸€èµ·ï¼Œå¹¶ä¸”æ¯ä¸ªå˜é‡çš„ç³»æ•°éƒ½ç”¨ä¸€ä¸ª**å‘é‡**è¡¨ç¤ºï¼Œå¦‚æŠŠ$x$çš„ç³»æ•°æ”¾åœ¨ä¸€èµ·ï¼Œå°±æ˜¯$\left[\begin{array}{l}1\\2\end{array}\right]$,ç±»ä¼¼çš„ï¼Œæˆ‘ä»¬æŠŠ$y$çš„ç³»æ•°ä¹Ÿæ”¾åœ¨ä¸€èµ·ï¼Œå¾—åˆ°$\left[\begin{array}{l}1\\4\end{array}\right]$ã€‚æœ€åï¼Œå¦‚æœæˆ‘ä»¬æŠŠæ‰€æœ‰æ•°å€¼éƒ½ç§»åˆ°ç­‰å¼çš„å·¦è¾¹ï¼Œä½¿å¾—æ–¹ç¨‹ç»„ä¸­æ¯ä¸ªç­‰å¼å³è¾¹éƒ½ä¸ºé›¶ï¼Œé‚£ä¹ˆç»“æœäº‹å®ä¸Šå¯ä»¥çœ‹æˆæ˜¯$1$çš„ç³»æ•°ï¼Œç§°ä¸ºå¸¸æ•°é¡¹ï¼Œå¦‚æœæˆ‘ä»¬ä¹Ÿæ”¾åœ¨ä¸€èµ·ï¼Œå¾—åˆ°$\left[\begin{array}{l}10\\28\end{array}\right]$.æœ€ç»ˆï¼Œæˆ‘ä»¬çš„æ–¹ç¨‹ç»„$(\ref{eq1})$å¯ä»¥ç”¨å‘é‡å½¢å¼è¡¨ç¤ºæˆ

$$
\left\{\begin{array}{l}
x+y=10\\

\end{array}\right.
\Rightarrow
\begin{bmatrix}
{1}\\
{2}\\
\end{bmatrix}
x+\begin{bmatrix}
{1}\\
{4}\\

{10}\\
{28}\\
\end{bmatrix}\label{eq*}
$$

å¦‚æœæˆ‘ä»¬å°†å‰é¢åŒ…å«æœªçŸ¥æ•°çš„ä¸¤ä¸ªå‘é‡æ‹¼æ¥åœ¨ä¸€èµ·ï¼Œå¯ä»¥å¾—åˆ°ä¸€ä¸ªäºŒç»´çš„æ•°è¡¨
$$\begin{bmatrix}
{1\quad1}\\
{2\quad4}\\
\end{bmatrix}$$
åƒè¿™æ ·çš„æ•°è¡¨æˆ‘ä»¬ä»¥åå°†å…¶ç§°ä¸º**çŸ©é˜µ**ï¼Œç”±äºåœ¨æ–¹ç¨‹ç»„ä¸­ï¼Œè¿™ä¸ªçŸ©é˜µè¡¨ç¤ºçš„æ˜¯æœªçŸ¥æ•°çš„ç³»æ•°ï¼Œæ•…ç§°ä¸ºç³»æ•°çŸ©é˜µï¼Œå¸¸ç”¨$\mathbf{A}$è¡¨ç¤º.ç»“æœå‘é‡ä¹Ÿç§°ä¸ºå¸¸æ•°å‘é‡ï¼Œé€šå¸¸è®°ä½œ$b$.

ä¸‹é¢æˆ‘ä»¬å°è¯•ä½¿ç”¨pythonè§£çº¿æ€§æ–¹ç¨‹ç»„ï¼Œ```Numpy```ä¸­å·²ç»å°è£…äº†æ±‚è§£çº¿æ€§æ–¹ç¨‹ç»„çš„å‡½æ•°ï¼Œæˆ‘ä»¬ä»…éœ€è¦ä¼ å…¥å¯¹åº”çš„ç³»æ•°çŸ©é˜µ$\mathbf{A}$ä»¥åŠå¸¸æ•°å‘é‡$b$ï¼Œç¨‹åºå°±ä¼šç®—å‡ºç›¸åº”çš„ç»“æœ.

````python
import numpy as np 
A = np.array([[1, 1],
              [2, 4]])     # å°†ç³»æ•°æ‰€æœ‰å‘é‡æ‹¼åœ¨ä¸€èµ·
b = np.array([10,
              28])  # å¸¸æ•°å‘é‡
x = np.linalg.solve(A,b)   # è§£çº¿æ€§æ–¹ç¨‹ç»„
print("çº¿æ€§æ–¹ç¨‹ç»„çš„è§£ä¸ºï¼š\n",x)
````

## 2.2 å‘é‡ç©ºé—´ã€çŸ©é˜µã€è¡Œåˆ—å¼ä»¥åŠèŒƒæ•°

<div class="alert alert-info" role="alert">
â“ ç„¶è€Œï¼ŒGingerå¸Œæœ›å¤šå…»å‡ ç§åŠ¨ç‰©ï¼Œè¿‘æ—¥ä»–åˆå¼•å…¥äº†å‡ åªå…¬é¸¡å’Œå‡ åªé¸­å­ï¼Œç°åœ¨é‡‡é›†åˆ°çš„æ•°æ®æ˜¯ä¸€å…±æœ‰14ä¸ªå¤´ï¼Œ40åªè„šï¼Œè¯·é—®é¸¡å…”é¸­å„æœ‰å‡ åªï¼Ÿ

å‡è®¾é¸­å­çš„æ•°ç›®æ˜¯$z$åªï¼Œæˆ‘ä»¬ç›´æ¥åˆ—å‡ºæ–¹ç¨‹ç»„
$$
\left\{\begin{array}{l}

æˆ‘ä»¬ç›´æ¥é‡‡ç”¨Pythonæ±‚è§£ï¼š

````python
A = np.array([[1, 1, 1],
              [2, 4, 2]])     # å°†ç³»æ•°æ‰€æœ‰å‘é‡æ‹¼åœ¨ä¸€èµ·
b = np.array([14,
              40])  # å¸¸æ•°å‘é‡
x = np.linalg.solve(A,b)   # è§£çº¿æ€§æ–¹ç¨‹ç»„
print("çº¿æ€§æ–¹ç¨‹ç»„çš„è§£ä¸ºï¼š\n",x)
````

<div class="alert alert-danger" role="alert">
è¿™ä¸ªä»£ç ä¸€å®šä¼šæŠ¥é”™ï¼å®ƒçš„æŠ¥é”™åŸå› è¯´ï¼Œä¸€å®šè¦æ±‚ç³»æ•°çŸ©é˜µæ˜¯ä¸ªæ–¹é˜µï¼Œè¿™æ˜¯å› ä¸ºè¦ç¡®å®šçš„æ±‚è§£ä¸€ä¸ªå¤šå…ƒæ–¹ç¨‹ç»„ï¼Œæœ‰å¤šå°‘ä¸ªæœªçŸ¥æ•°ï¼Œå°±éœ€è¦æœ‰å¤šå°‘ä¸ªæ–¹ç¨‹ï¼æ‰€ä»¥ç³»æ•°çŸ©é˜µä¸€å®šæ˜¯ä¸ªæ–¹é˜µ.

é‚£ä¹ˆï¼Œå¦‚æœæˆ‘ä»¬åŠ å…¥åŠ¨ç‰©çœ¼ç›çš„ç»Ÿè®¡æ•°æ®ï¼Œæ–¹ç¨‹ç»„å¯ä»¥ä¸ºï¼š

$$
\left\{\begin{array}{l}
x + y + z = 14 \\

2x + 2y + 2z = 28

$$
æˆ‘ä»¬ç°åœ¨å†å»çœ‹æ¯ä¸ªå˜é‡çš„ç³»æ•°å‘é‡ï¼Œå®é™…ä¸Šè¢«æ¨å¹¿åˆ°äº†ä¸‰é˜¶ï¼Œå½“å®é™…ç”Ÿæ´»ä¸­åŠ¨ç‰©è¶Šæ¥è¶Šå¤šæ—¶ï¼Œæˆ‘ä»¬éœ€è¦å¼•å…¥çš„å˜é‡ä¹Ÿè¶Šæ¥è¶Šå¤šï¼Œæ‰€ä»¥æˆ‘ä»¬éœ€è¦å¼•å…¥æ›´é«˜ç»´åº¦çš„å‘é‡æ¥è¡¨è¾¾æˆ‘ä»¬çš„é—®é¢˜.ä¸€èˆ¬åœ°ï¼Œæˆ‘ä»¬ç ”ç©¶çš„é—®é¢˜ï¼Œå˜é‡çš„ä¸ªæ•°éƒ½æ˜¯æœ‰é™çš„ï¼Œæ‰€ä»¥æˆ‘ä»¬ä¸€èˆ¬ç ”ç©¶çš„å‘é‡éƒ½æ˜¯æœ‰é™ç»´çš„å‘é‡ï¼Œä¸€èˆ¬ç§°ä¸º**nç»´å‘é‡**.

ç»§ç»­ä½¿ç”¨pythonæ±‚è§£$\eqref{eq5}$çš„é—®é¢˜

````python
A = np.array([[1, 1, 1],
              [2, 4, 2],
              [2, 2, 2]])     # å°†ç³»æ•°æ‰€æœ‰å‘é‡æ‹¼åœ¨ä¸€èµ·
b = np.array([14,
              40,
              28])  # å¸¸æ•°å‘é‡
x = np.linalg.solve(A,b)   # è§£çº¿æ€§æ–¹ç¨‹ç»„
print("çº¿æ€§æ–¹ç¨‹ç»„çš„è§£ä¸ºï¼š\n",x)
````

å‘ç°äº‹å®ä¸Šè¿˜æ˜¯æŠ¥é”™ï¼Œåªä¸è¿‡è¿™æ¬¡æŠ¥é”™çš„åŸå› æ˜¯```Singular matrix```ï¼Œæ„æ€æ˜¯ç³»æ•°çŸ©é˜µæ˜¯**å¥‡å¼‚çš„**ï¼Œæˆ–è€…è¯´æ˜¯**é€€åŒ–çš„**.å¦‚æœè§‚å¯Ÿæ–¹ç¨‹ç»„$\eqref{eq3}$ï¼Œä½ ä¼šå‘ç°å…¶å®ç¬¬ä¸‰ä¸ªå¼å­æ˜¯ç¬¬ä¸€ä¸ªå¼å­çš„2å€å¯¼å‡ºçš„ï¼Œå¦‚æœæˆ‘ä»¬é‡‡ç”¨é«˜æ–¯æ¶ˆå…ƒæ³•æ±‚è§£æ–¹ç¨‹ï¼Œé‚£ä¹ˆå°†å¼å­1çš„-2å€åŠ åˆ°ç¬¬3ä¸ªå¼å­ï¼Œç¬¬ä¸‰ä¸ªå¼å­å°±ä¸º0äº†.è¿™è¯´æ˜æˆ‘ä»¬æ–°å¢çš„è¿™ä¸ªå¼å­å¯¹äºè§£æ–¹ç¨‹å®é™…ä¸Šèµ·ä¸åˆ°ä»»ä½•ä½œç”¨çš„ï¼Œå› ä¸ºå®ƒçš„ä¿¡æ¯å·²ç»è¢«å›Šæ‹¬åœ¨ç¬¬ä¸€ä¸ªå¼å­ä¸­äº†.

äºæ˜¯ç°åœ¨æˆ‘ä»¬éœ€è¦è§£å†³çš„é—®é¢˜æœ‰ä¸¤ä¸ªï¼š
1. å½“æ±‚è§£nä¸ªæœªçŸ¥æ•°æ—¶ï¼Œæ–¹ç¨‹ç»„è¦æ»¡è¶³ä»€ä¹ˆæ¡ä»¶ï¼Œæ‰æ˜¯æœ‰è§£çš„ï¼Ÿ
2. å½“æˆ‘ä»¬å°†å‘é‡æ¨å¹¿åˆ°é«˜ç»´ä¹‹åï¼Œå‘é‡ä¹‹é—´çš„è®¡ç®—æ˜¯æ»¡è¶³çš„å—ï¼Ÿ

### 2.2.1 å‘é‡çš„è¿ç®—æ³•åˆ™
ä»ä¸Šé¢çš„æ–¹ç¨‹ç»„æ”¹å†™æˆå‘é‡çš„è¡¨è¾¾å½¢å¼æ—¶ï¼Œå¤§å®¶å¿ƒä¸­ä¹Ÿè®¸ä¼šæƒ³é—®ä¸€ä¸ªé—®é¢˜ï¼šè¿™ä¸¤ç§å†™æ³•æ˜¯ç­‰ä»·çš„å—ï¼Ÿæˆ‘ä»¬çœ‹$\eqref{eq*}$çš„è®¡ç®—è¿‡ç¨‹ï¼š
$$

{1}\\

\end{bmatrix}

+\begin{bmatrix}

ç”±äºè¦ä¿è¯å¯¹åº”çš„åˆ†é‡ç›¸ç­‰ï¼Œæ‰€ä»¥è‡ªç„¶æˆ‘ä»¬å°±å¯ä»¥æ¢å¤å‡ºæ–¹ç¨‹$\eqref{eq1}$.
ä¸Šé¢è¿™ä¸ªè¿‡ç¨‹ï¼Œå‘é‡çš„è®¡ç®—æœ‰ä¸¤ç§ï¼š
- ä¸€ä¸ªæ•°ä¹˜ä¸€ä¸ªå‘é‡ï¼›
- ä¸€ä¸ªå‘é‡åŠ ä¸€ä¸ªå‘é‡ï¼›
æ˜¾ç„¶ï¼Œå½“å˜é‡ä¸ªæ•°å˜å¤šä¹‹åï¼Œä¹Ÿä¼šæ»¡è¶³ä¸Šé¢è¿™ä¸¤ç§è¿ç®—æ³•åˆ™ï¼Œå› æ­¤æˆ‘ä»¬æ¥ä¸‹æ¥æ­£å¼å¼•å…¥å‘é‡çš„åŸºæœ¬è¿ç®—æ³•åˆ™ï¼š

ç»™å®š$n$ç»´å‘é‡
$$

$$
- å‘é‡çš„åŠ æ³•å®šä¹‰ä¸ºï¼š
$$
x+y = \begin{bmatrix}
{x_1 + y_1}\\
{x_2 + y_2}\\
{\vdots}\\
{x_n + y_n}
\end{bmatrix},
$$
å³ä¸¤ä¸ªå‘é‡å¯¹åº”çš„åˆ†é‡ç›¸åŠ .

- å‘é‡çš„æ•°ä¹˜å®šä¹‰ä¸ºï¼š
å¯¹äº$k \in \mathbf{R}$,
$$

å³æ¯ä¸ªåˆ†é‡éƒ½ä¹˜ä¸Š$k$.

å…³äºå‘é‡ä¸å‘é‡çš„ä¹˜æ³•ï¼Œåœ¨åç»­çš„å­¦ä¹ ä¸­ï¼Œæˆ‘ä»¬å†è¿›ä¸€æ­¥ä»‹ç».æˆ‘ä»¬çº¦å®šä¸€ä¸ªä½¿ç”¨ä¹ æƒ¯ï¼Œæˆ‘ä»¬é»˜è®¤æ‰€æœ‰çš„å‘é‡éƒ½é‡‡ç”¨åˆ—å‘é‡çš„å½¢å¼ï¼Œå¦‚ä¸Šé¢æ‰€ç¤º.å½“æˆ‘ä»¬åœ¨æ–‡ä¸­ä¹¦å†™æ—¶ï¼Œä¸ºäº†æ’ç‰ˆä¼˜ç¾ï¼Œé€šå¸¸å°†å‘é‡ç”±åˆ—å‘é‡è½¬æˆè¡Œå‘é‡è¿›è¡Œä¹¦å†™ï¼Œå‘é‡çš„è¡Œåˆ—äº¤æ¢çš„ç§°ä¸ºå‘é‡çš„**è½¬ç½®**ï¼Œè®°åš$x^T = (x_1, x_2, \cdots, x_n)$.æˆ‘ä»¬ä½¿ç”¨```Numpy```å®ç°å‘é‡ä¹‹é—´çš„åŸºæœ¬è®¡ç®—.

````python
import numpy as np
# ç”Ÿæˆå‘é‡
x = np.array([1, 2, 3]) # arrayé»˜è®¤å¦‚æœåªæœ‰ä¸€åˆ—ï¼Œå°±æ˜¯ä¸€ä¸ªå‘é‡
y = np.array([4, 5, 6])
print("x={},y={}".format(x, y))
print("xçš„ç»´åº¦ä¸º{}".format(x.shape)) # shapeå‡½æ•°ç”¨äºæ˜¾ç¤ºå‘é‡çš„ç»´åº¦ï¼Œå¦‚æœæ˜¯å‘é‡é»˜è®¤åªæœ‰ä¸€ç»´ï¼Œç»´åº¦æ˜¾ç¤ºä¸º(dim,)

# å‘é‡åŠ æ³•
print("x+y = {}".format(x + y))

# å‘é‡æ•°ä¹˜
k = 3
print("kx = {}".format(k*x))

print("3x+2y ={} ".format(3*x + 2*y))
````
